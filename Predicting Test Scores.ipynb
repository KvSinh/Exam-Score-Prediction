### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Importing Dataset

```python
dataset= pd.read_csv('student_scores - student_scores.csv')
X= dataset.iloc[:,0].values
Y=dataset.iloc[:,-1].values
```

### Splitting data into Train and Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.2, random_state=0)
```

### Fitting Regression on Training Data

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(X_train.reshape(len(X_train),1), Y_train.reshape(len(Y_train),1))
```

### Predicting Result of a student which studies for 9.5 hours

```python
print(lr.predict([[9.25]]))
```

### Calculating fit using R squared

```python
Y_pred= lr.predict(X_test.reshape(len(X_test),1))
```

```python
from sklearn.metrics import r2_score
print(r2_score(Y_pred, Y_test))
```

### Plotting Result based on Training Data

```python
plt.scatter(X_train, Y_train)
plt.plot(X_train, lr.predict(X_train.reshape(len(X_train),1)), color="red")
plt.title("Predicting Percentage of Student based on hours of study (Training)")
plt.xlabel("Numbers of Hours of Study")
plt.ylabel("Percentage Score")
plt.show()
```

### Plotting Result based on Test Data

```python
plt.scatter(X_test, Y_test)
plt.plot(X_train, lr.predict(X_train.reshape(len(X_train),1)), color="red")
plt.title("Predicting Percentage of Student based on hours of study (Test)")
plt.xlabel("Numbers of Hours of Study")
plt.ylabel("Percentage Score")
plt.show()
```
